Средняя энтропия AE

```python
import matplotlib.pyplot as plt
import numpy as np

# Данные первой выборки
data1 = [0.017499, 0.110607, 0.025061, 0.088142, 0.818982, 0.247428, 0.021587, 0.1131960, 0.549410, 0.196831]

# Данные второй выборки
data2 = [0.549341, 0.245357, 0.415522, 0.478520, 0.417813, 0.106349, 0.385584, 0.222451, 0.369530, 0.680255]

# Объединяем данные для построения одного графика
data = [data1, data2]

# Построение графика "box plot"
plt.boxplot(data)
   
# Добавление точек для каждого значения
plt.plot([1] * len(data1), data1, 'ro', label='AE-1')
plt.plot([2] * len(data2), data2, 'bo', label='AE-2')
   
# Настройка осей и меток
plt.xticks([1, 2], ['AE - Mini-AES', 'AE - White-box Mini-AES'])
plt.ylabel('Средняя энтропия')

# Добавление линий для значений Q1, Q2 и Q3
plt.axhline(y=np.percentile(data1, 50), linestyle='--', color='r', label='Q2 AE-1 (Медиана)')
plt.axhline(y=np.percentile(data2, 50), linestyle='--', color='b', label='Q2 AE-2 (Медиана)')

# Добавление точного числового значения медианы
plt.annotate(f'{np.percentile(data1, 50):.4f}', xy=(1, np.percentile(data1, 50)), xytext=(-50, 10),
             textcoords='offset points', ha='right', va='center', color='r')
plt.annotate(f'{np.percentile(data2, 50):.4f}', xy=(2, np.percentile(data2, 50)), xytext=(-263, 10),
             textcoords='offset points', ha='left', va='center', color='b')


# Добавление легенды
plt.legend()

# Отображение графика
plt.show()
```
